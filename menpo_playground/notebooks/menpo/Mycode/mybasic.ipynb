```python
%matplotlib inline
import numpy as np
import menpo.io as mio
from menpo.image import Image

#Import image. 
#Print image type, instance, pixels shape, n_channels, shape, width, height, semantic labels, center, dimention, 
#Print n_pixels, n_elements
lenna = mio.import_builtin_asset( 'lenna.png' )
print( 'Lenna is a {}'.format( type( lenna ) ) )
print( "Lenna 'is an' Image: {}".format( isinstance( lenna, Image) ) )
print( 'Lenna shape: {}'.format( lenna.pixels.shape ) )
print( 'The number of channels in Lenna is {}'.format( lenna.pixels.shape[ 0 ] ) )
print( 'n_channels for Lenna is {}'.format( lenna.n_channels) )
print( 'Lenna has a shape of {}'. format( lenna.shape ) )
print( 'Lenna has W:{} H:{}'.format( lenna.width, lenna.height ) )
print( lenna )
print( 'The Centre of Lenna is {}'. format( lenna.centre( ) ) )# ( Y, X )
print( 'Lenna n_dims: {}'.format( lenna.n_dims ) )
print( 'Lenna n_pixels: {}'.format( lenna.n_pixels ) ) #512*512
print( 'Lenna n_elements: {}'.format( lenna.n_elements ) ) # 3*512*512




```

```python
#View Lenna( figure_size with foot unit)
lenna.view( render_axes = True, axes_font_weight = 'bold', alpha = 0.8, figure_size = ( 7, 5 ) );
```

```python
lenna.view( channels = [ 0, 1, 2 ] );
```

```python
#Crop: return the corpped image
lenna_copped = lenna.crop( ( 0, 0 ), lenna.centre( ) )
lenna_copped.view( figure_size = ( 5, 5 ) );
```

```python
# Resceal
lenna_half = lenna.rescale( 0.5 )
print( lenna_half )
```

```python
#View landmark
lenna.view_landmarks( group = 'LJSON' );
```

```python
lenna2 = lenna.crop_to_landmarks( boundary = 20 )
lenna2.view_landmarks( group = 'LJSON', marker_edge_colour = 'k', marker_face_colour = 'w', marker_edge_width = 2 );
```

```python
#Masked Image
lenna_masked = lenna.as_masked() #return a image with mask
print( lenna_masked.mask )
lenna_masked.mask.view( new_figure=True, figure_size=(6, 4), render_axes=True );
lenna_masked.view( new_figure = True );
```

```python
#Constrain mask to landmarks
lenna_masked_constrained = lenna_masked.constrain_mask_to_landmarks()
lenna_masked_constrained.mask.view();
lenna_masked_constrained.view( new_figure=True );
lenna_masked_constrained.view( masked = False, new_figure = True );
lenna_masked_constrained.view_landmarks( new_figure=True );
```
